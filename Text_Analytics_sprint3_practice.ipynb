#Task-4
Clean the data and save it as a single dataframe.

```python
import re
```

```python
def clean_text(text):
    """
    Return clean version of the text

    """
    # Remove all non-letters and non-spaces except for hyphens and digits
    text = re.sub("[^0-9A-Za-z\- ]+", " ", text)
    # Remove all numbers except those attached to a word
    text = re.sub("(?<!\w)\d+", "", text)
    # Remove all hyphens except between two words
    text = re.sub(".(?!\w)|(?<!\w).", "", text)
    # Remove multiple spaces and lowercase everything
    text = " ".join(text.split())
    text = text.lower()

    return text
```

```python
df['text']=df['text'].apply(clean_text)
#df['target']=df['target'].apply(clean_text)
```

#Task-6
Perform multi-label classification with fasttext.

```python
with open('concept_train.txt','w+', encoding="utf-8") as train_file:
    for i in train_df.index:
        labels = ''
        for target in train_df['target'][i]:
            labels = labels + '__label__' + target.replace(' ', '_') + ' '
        line = labels + ' '+train_df['text'][i]
        train_file.write(line+'\n')

with open('concept_test.txt','w+', encoding="utf-8") as test_file:
    for i in test_df.index:
        labels = ''
        for target in test_df['target'][i]:
            labels = labels + '__label__'+ target.replace(' ', '_') + ' '
        line = labels + ' '+test_df['text'][i]
        test_file.write(line+'\n')
```

#Task-7
Calculate and display the f1-score in a percentage

```python
for i in [20,25,30]:
    model_sup = fasttext.train_supervised(input='concept_train.txt', lr=0.5, epoch=i, wordNgrams=3, bucket=20000)
    test_num, precision, recall = model_sup.test('concept_test.txt', k=1, threshold=0.1)
    f_score = (2*precision*recall)/(precision+recall)

for i in zip(model_sup.words, model_sup.labels):
    print(i)
```
